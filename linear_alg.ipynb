## Algebra Recap

```python
import numpy as np
```

```python
def vector_vector_multi(u, v):
    assert u.shape[0] == v.shape[0]

    sum = np.sum(u * v)
    return sum
```

```python
def vector_vector_multi_ext(u, v):
    assert u.shape[0] == v.shape[0]
    res = 0

    for i in range(u.shape[0]):
        res += u[i] * v[i]
    return res
```

```python
u = np.array([2, 4, 5, 6])
v = np.array([1, 0, 0, 2])
vector_vector_multi(u, v)
vector_vector_multi_ext(np.array([2, 4, 5, 6]), np.array([1, 0, 0, 2]))

u.dot(v)
```

# Matrix Vector Multiplication


```python
A = np.array([
    [2, 4, 5, 6],
    [1, 2, 1, 2],
    [3, 1, 2, 1]
])

v = np.array([1, 0.5, 2, 1])

def matrix_vector_multiplication(U, v):
    assert U.shape[1] == v.shape[0]

    n_rows = U.shape[0]
    result = np.zeros(n_rows)

    for i in range(n_rows):
        Uiv = vector_vector_multi(U[i], v)
        #print(Uiv)
        result[i] = Uiv
    return result

print(matrix_vector_multiplication(A, v))
```

## Matrix-Matrix multiplication

```python
A = np.array([
    [2, 4, 5, 6],
    [1, 2, 1, 2],
    [3, 1, 2, 1]
])

B = np.array([
    [1, 1, 2],
    [0, 0.5, 1],
    [0, 2, 1],
    [2, 1, 0]
])
def matrix_matrix_mult(U, V):
    assert U.shape[1] == V.shape[0]
    n_rows = U.shape[0]
    n_cols = V.shape[1]
    resp = np.zeros((n_rows, n_cols))

    for i in range(B.shape[1]):
        vi = B[:, i]
        Uvi = matrix_vector_multiplication(U, vi)
        resp[:, i] = Uvi

    return resp

print(matrix_matrix_mult(A, B))
```
